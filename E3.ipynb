Q1 [nyc_weather.csv](https://github.com/codebasics/data-structures-algorithms-python/blob/master/data_structures/4_HashTable_2_Collisions/Solution/nyc_weather.csv) contains new york city weather for first few days in the month of January. Write a program that can answer following,\
What was the average temperature in first week of Jan\
What was the maximum temperature in first 10 days of Jan\
Figure out data structure that is best for this problem

````python
arr = []
with open("nyc_weather.csv") as f:
    for line in f:
        arr.append(line.strip().split(","))
print(arr)

# What was the average temperature in first week of January?

total_temp = 0
for i in range(1, 8):
    total_temp += int(arr[i][1])
avg_temp = total_temp / 7
print(avg_temp)

# What was the maximum temperature in 10 days of January?

max_temp = int(arr[1][1])
for i in range(2, 11):
    temp = int(arr[i][1])
    if temp > max_temp:
        max_temp = temp
print(max_temp)
````

Q2 [nyc_weather.csv](https://github.com/codebasics/data-structures-algorithms-python/blob/master/data_structures/4_HashTable_2_Collisions/Solution/nyc_weather.csv) contains new york city weather for first few days in the month of January. Write a program that can answer following,\
What was the temperature on Jan 9?\
What was the temperature on Jan 4?\
Figure out data structure that is best for this problem

````python
dict = {}
with open("nyc_weather.csv") as f:
    for line in f:
        date, temp = line.strip().split(",")
        try:
            temp = int(temp)
        except ValueError:
            continue
        dict[date] = int(temp)
print(dict)

#What was the temperature on Jan 9?

print(dict["Jan 9"])

# What was the temperature on Jan 4?

print(dict["Jan 4"])
````

Q3 [poem.txt](https://github.com/codebasics/data-structures-algorithms-python/blob/master/data_structures/4_HashTable_2_Collisions/Solution/poem.txt) Contains famous poem "Road not taken" by poet Robert Frost. You have to read this file in python and print every word and its count as show below. Think about the best data structure that you can use to solve this problem and figure out why you selected that specific data structure.  
```python
 'diverged': 2,
 'in': 3,
 'I': 8
 ```

````python
dict = {}
with open("poem.txt") as f:
    for line in f:
        words = line.strip().split(" ")
        for word in words:
            word = word.strip(".,!?;:")
            if word in dict:
                dict[word] += 1
            else:
                dict[word] = 1
print(dict)
````

Q4 Implement hash table where collisions are handled using linear probing. We learnt about linear probing in the video tutorial. Take the hash table implementation that uses chaining and modify methods to use linear probing. Keep MAX size of arr in hashtable as 10.

````python
class HashTable:
    def __init__(self):
        self.MAX = 10
        self.arr = [[] for i in range(self.MAX)]

    def get_hash(self, key):
        hash = 0
        for char in key:
            hash += ord(char)
        return hash % self.MAX
    
    def __getitem__(self, key):
        index = self.get_hash(key)
        initial_h = index
        while self.arr[index]:
            if self.arr[index][0][0] == key:
                return self.arr[index][0][1]
            index = (index + 1) % self.MAX
            if index == initial_h:
                break
        return None
            
    def __setitem__(self, key, value):
        index = self.get_hash(key)
        initial_h = index
        while self.arr[index] and self.arr[index][0][0] != key:
            index = (index + 1) % self.MAX
            if index == initial_h:
                raise Exception("Hash table is full")
            
        self.arr[index] = [(key, value)]

    def __delitem__(self, key):
        index = self.get_hash(key)
        initial_h = index
        while self.arr[index]:
            if self.arr[index][0][0] == key:
                self.arr[index] = []
                return
            index = (index + 1) % self.MAX
            if index == initial_h:
                break

h = HashTable()
h["march 6"] = 130
h["march 7"] = 131
h["march 8"] = 132
h["march 17"] = 133
````
