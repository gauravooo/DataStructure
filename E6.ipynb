Extent tree class built in our main tutorial so that it takes name and designation in data part of TreeNode class. Now extend print_tree function such that it can print either name tree, designation tree or name and designation tree. As shown below,

```python
if __name__ == '__main__':

    root_node.print_tree("name") # prints only name hierarchy
    root_node.print_tree("designation") # prints only designation hierarchy
    root_node.print_tree("both") # prints both (name and designation) hierarchy
```

````python
# Extent tree class built in our main tutorial so that it takes name and designation in data part of TreeNode class. Now extend print_tree function such that it can print either name tree, designation tree or name and designation tree.
class TreeNode:
    def __init__(self, name, designation):
        self.name = name
        self.designation = designation
        self.children = []
        self.parent = None

    def add_child(self, child_node):
        child_node.parent = self
        self.children.append(child_node)

    def get_level(self):
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent
        return level

    def print_tree(self, property_name):
        if property_name == "name":
            value = self.name
        elif property_name == "designation":
            value = self.designation
        elif property_name == "both":
            value = f"{self.name} ({self.designation})"
        else:
            raise ValueError("Invalid property name. Use 'name', 'designation', or 'both'.")

        spaces = ' ' * self.get_level() * 3
        prefix = spaces + "|__" if self.parent else ""
        print(prefix + value)
        for child in self.children:
            child.print_tree(property_name)
def build_management_tree():
    ceo = TreeNode("Nilupul", "CEO")

    cto = TreeNode("Chinmay", "CTO")
    hr_head = TreeNode("Vishwa", "HR Head")
    cfo = TreeNode("Gels", "CFO")

    ceo.add_child(cto)
    ceo.add_child(hr_head)
    ceo.add_child(cfo)

    in_engineering = TreeNode("Aamir", "Infrastructure Head")
    app_engineering = TreeNode("Dhaval", "Application Head")

    cto.add_child(in_engineering)
    cto.add_child(app_engineering)

    in_engineering.add_child(TreeNode("Peter", "Infrastructure Engineer"))
    in_engineering.add_child(TreeNode("Waqas", "Infrastructure Engineer"))

    app_engineering.add_child(TreeNode("Sara", "Application Engineer"))
    app_engineering.add_child(TreeNode("Anuj", "Application Engineer"))

    hr_head.add_child(TreeNode("Peter", "Recruitment Manager"))
    hr_head.add_child(TreeNode("Waqas", "Policy Manager"))

    cfo.add_child(TreeNode("Sara", "Accountant"))
    cfo.add_child(TreeNode("Anuj", "Financial Analyst"))

    return ceo
if __name__ == '__main__':
    root_node = build_management_tree()
    print("Printing only names:")
    root_node.print_tree("name")
    print("\nPrinting only designations:")
    root_node.print_tree("designation")
    print("\nPrinting both names and designations:")
    root_node.print_tree("both")

````

Q2 Build below location tree using TreeNode class  
  
Now modify print_tree method to take tree level as input. And that should print tree only upto that level as shown below,  
  

````python
# build location tree ,Now modify print_tree method to take tree level as input. And that should print tree only upto that level as shown below,
class TreeNode:
    def __init__(self, name):
        self.name = name
        self.children = []
        self.parent = None

    def add_child(self, child_node):
        child_node.parent = self
        self.children.append(child_node)

    def get_level(self):
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent
        return level

    def print_tree(self, max_level):
        if self.get_level() > max_level:
            return

        spaces = ' ' * self.get_level() * 3
        prefix = spaces + "|__" if self.parent else ""
        print(prefix + self.name)
        for child in self.children:
            child.print_tree(max_level)

def build_location_tree():
    world = TreeNode("Global")

    india = TreeNode("India")
    usa = TreeNode("USA")

    world.add_child(india)
    world.add_child(usa)

    gujarat = TreeNode("Gujarat")
    karnataka = TreeNode("Karnataka")

    india.add_child(gujarat)
    india.add_child(karnataka)

    baroda = TreeNode("Baroda")
    ahmedabad = TreeNode("Ahmedabad")

    gujarat.add_child(baroda)
    gujarat.add_child(ahmedabad)

    bangalore = TreeNode("Bangalore")
    mysore = TreeNode("Mysore")

    karnataka.add_child(bangalore)
    karnataka.add_child(mysore)

    california = TreeNode("California")
    new_jersey = TreeNode("New Jersey")

    usa.add_child(california)
    usa.add_child(new_jersey)

    san_francisco = TreeNode("San Francisco")
    mountain_view = TreeNode("Mountain View")
    palo_alto = TreeNode("Palo Alto")

    california.add_child(san_francisco)
    california.add_child(mountain_view)
    california.add_child(palo_alto)

    princeton = TreeNode("Princeton")
    trenton = TreeNode("Trenton")

    new_jersey.add_child(princeton)
    new_jersey.add_child(trenton)

    return world
if __name__ == '__main__':
    root_node = build_location_tree()
    root_node.print_tree(1)  # Print tree up to level 1
    print("\n")
    root_node.print_tree(2)  # Print tree up to level 2
    print("\n")
    root_node.print_tree(3)  # Print tree up to level 3
    
````
