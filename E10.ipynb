Q2.Find index of all the occurances of a number from sorted list  
```python
numbers = [1,4,6,9,11,15,15,15,17,21,34,34,56]
number_to_find = 15  
```
This should return 5,6,7 as indices containing number 15 in the array

````python
#return all index of target occuring multiple times in sorted array
def find_all_occurrences(arr, target):
    left, right = 0, len(arr) - 1
    result = []

    while left <= right:
        mid = left + (right - left) // 2

        if arr[mid] == target:
            # Found the target, now find all occurrences
            # Check to the left of mid
            i = mid
            while i >= 0 and arr[i] == target:
                result.append(i)
                i -= 1
            
            # Check to the right of mid
            i = mid + 1
            while i < len(arr) and arr[i] == target:
                result.append(i)
                i += 1
            
            break  # Exit after finding all occurrences
        elif arr[mid] < target:
            left = mid + 1   
        else:
            right = mid - 1
    return sorted(result)
# Example usage
numbers = [1,4,6,9,11,15,15,15,17,21,34,34,56]
target = 15
print(find_all_occurrences(numbers, target))  # Output: [5, 6, 7]
````
