Design a food ordering system where your python program will run two threads,

Place Order: This thread will be placing an order and inserting that into a queue. This thread places new order every 0.5 second. (hint: use time.sleep(0.5) function)
Serve Order: This thread will server the order. All you need to do is pop the order out of the queue and print it. This thread serves an order every 2 seconds. Also start this thread 1 second after place order thread is started.

Pass following list as an argument to place order thread,
```python
orders = ['pizza','samosa','pasta','biryani','burger']
```
This problem is a producer,consumer problem where place_order thread is producing orders whereas server_order thread is consuming the food orders. Use Queue class implemented in a video tutorial.

````python
# Food ordering system using a queue running two threads: Place Order and Serve Order
import threading
import time
class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return not self.items

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if self.is_empty():
            raise IndexError("Dequeue from an empty queue")
        return self.items.pop(0)

    def size(self):
        return len(self.items)
    
class FoodOrderingSystem:
    def __init__(self):
        self.order_queue = Queue()
        self.lock = threading.Lock()
        self.orders_done = False

    def place_order(self, order):
        with self.lock:
            self.order_queue.enqueue(order)
            print(f"Order placed: {order}")

    def set_done(self):
        with self.lock:
            self.orders_done = True

    def serve_orders(self):
        time.sleep(1)
        while True:
            with self.lock:
                if not self.order_queue.is_empty():
                    order = self.order_queue.dequeue()
                    print(f"Order served: {order}")
                elif self.orders_done:
                    print("No more orders to serve.")
                    break
                else:
                    order = None

            if order is not None:
                time.sleep(2)
            else:
                time.sleep(0.5)

        print("Serve thread exiting.")

def main():
    food_system = FoodOrderingSystem()

    # Thread for placing orders
    def place_orders():
        orders = ['pizza','samosa','pasta','biryani','burger']
        for order in orders:
            food_system.place_order(order)
            time.sleep(0.5)
        food_system.set_done()

    # Thread for serving orders
    serve_thread = threading.Thread(target=food_system.serve_orders)

    # Create threads
    place_thread = threading.Thread(target=place_orders)

    # Start threads
    place_thread.start()
    serve_thread.start()

    # Wait for both threads to finish
    place_thread.join()
    serve_thread.join()

if __name__ == "__main__":
    main()



````
