# Data structure tutorial exercise: Stack

Q1 Write a function in python that can reverse a string using stack data structure. Use Stack class from the tutorial.  
```python
reverse_string("We will conquere COVID-19") should return "91-DIVOC ereuqnoc lliw eW"
```

````python
class stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        else:
            raise IndexError("Stack is empty")

    def is_empty(self):
        return len(self.stack) == 0

    def reverse_string(self, s):
        for char in s:
            self.push(char)
        reversed_string = ''
        while not self.is_empty():
            reversed_string += self.pop()
        return reversed_string

if __name__ == "__main__":
    stack_instance = stack()
    s = input("Enter a string to reverse: ")
    reversed_s = stack_instance.reverse_string(s)
    print("Reversed string:", reversed_s)
````
